Calculating Histogram

cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])

images : it is the source image of type uint8 or float32 represented as “[img]”.
    
channels : it is the index of channel for which we calculate histogram.
For grayscale image, its value is [0] and
color image, you can pass [0], [1] or [2] to calculate histogram of blue, green or red channel respectively.

mask : mask image. To find histogram of full image, it is given as “None”.
    
histSize : this represents our BIN count. For full scale, we pass [256].

ranges : this is our RANGE. Normally, it is [0,256].

```python
# importing required libraries of opencv
import cv2

# importing library for plotting
from matplotlib import pyplot as plt

# reads an input image
img = cv2.imread("home.jpg",0)

# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

[imageObj]: This parameter is the source image. It's passed as a list, where imageObj is the image you want to calculate the histogram for.



[0]: This specifies the channel for which you want to compute the histogram. In this case, [0] indicates that you're calculating the histogram for the first channel of the image. For grayscale images, there's only one channel (0-indexed), but for color images, you'd specify the channel you're interested in (e.g., [0] for blue channel, [1] for green, [2] for red in an RGB image).



None: This is an optional mask. If provided, it indicates a region in the image for which you want to calculate the histogram. If None is passed, the full image is considered.



[256]: This parameter specifies the number of bins for the histogram. In this case, it's set to [256], which means the histogram will have 256 bins. This is often used to divide the pixel values of the image into 256 intervals or bins and count the number of pixels in each bin.



[0, 256]: This is the range of pixel values you want to consider for the histogram. For example, [0, 256] means the full range of pixel values from 0 to 255 (256 values in total) will be considered. This range determines the values that will be binned into the specified number of bins.



In summary, the cv2.calcHist() function, in this case, calculates the histogram for the specified image (imageObj) in the first channel (assuming it's a grayscale image) with 256 bins, considering the full range of pixel values from 0 to 255.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("home.jpg")
#cv2.imshow("Sweet Home", img)
plt.imshow(img)
plt.show()
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
#import necessary packages
import cv2
import matplotlib.pyplot as plt

#load image
imageObj = cv2.imread('home.jpg')
#to avoid grid lines
plt.axis("off")
plt.title("Original Image")
plt.imshow(cv2.cvtColor(imageObj, cv2.COLOR_BGR2RGB))
plt.show()

#Get RGB data from image
blue_color = cv2.calcHist([imageObj],[0],None,[256],[0,256])
red_color = cv2.calcHist([imageObj],[1],None,[256],[0,256])
green_color = cv2.calcHist([imageObj],[2],None,[256],[0,256])

#Seperate Histograms for each color
plt.subplot(3,1,1)
plt.title("Histogram of Blue")
plt.hist(blue_color, color='blue')

plt.subplot(3,1,2)
plt.title("Histogram of Red")
plt.hist(red_color, color='red')

plt.subplot(3,1,3)
plt.title("Histogram of Green")
plt.hist(green_color, color='green')

#for clearer view
plt.tight_layout()
plt.show()

#combined histogram
plt.title("Histogram of all RGB Colors")
plt.hist(blue_color, color="blue")
plt.hist(red_color, color='red')
plt.hist(green_color, color='green')
plt.show()
```
