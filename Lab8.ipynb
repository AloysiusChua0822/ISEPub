```python
def translate_point(x, y, translate_x, translate_y):
    new_x = x + translate_x
    new_y = y + translate_y

    return new_x, new_y

x = float(input("Enter the X-coordinate of the point: "))
y = float(input("Enter the Y-coordinate of the point: "))

translate_x = float(input("Enter the translation along the X-axis: "))
translate_y = float(input("Enter the translation along the Y-axis: "))

new_x, new_y = translate_point(x, y, translate_x, translate_y)

print(f"Original Point: ({x}, {y})")
print(f"Translated Point: ({new_x},Â {new_y})")
```

```python
import matplotlib.pyplot as plt

def draw_line(x1, y1, x2, y2):
    # Plotting the points
    plt.plot([x1, x2], [y1, y2], marker='o')


    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Line between two points')


    plt.show()

# for input coord
x1 = float(input("Enter the X-coordinate of the first point: "))
y1 = float(input("Enter the Y-coordinate of the first point: "))
x2 = float(input("Enter the X-coordinate of the second point: "))
y2 = float(input("Enter the Y-coordinate of the second point: "))


draw_line(x1,y1,x2,y2)
```
